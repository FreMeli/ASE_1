###Calculado la tasa nacional de homicidios

```python
upload1 = files.upload()
```

```python
df_P=pd.read_csv(io.BytesIO(upload1["pob_ini_proyecciones.csv"]), header=0, encoding="ISO-8859-1")
df_P #Para ver las primeras 10 filas
```

```python
df_P.info()
```

```python
 pepe = df_P[df_P["AÑO"] >= 2000]
 pepe = pepe[pepe["AÑO"] <= 2021]
 pepe
```

```python
columns = ["AÑO", "ENTIDAD", "HOMBRES", "MUJERES"]
mtrx = []

for ano in pepe["AÑO"].unique():
  values_ano = pepe[pepe["AÑO"] == ano]
  for estado in values_ano["ENTIDAD"].unique():
    tmp = []
    tmp.append(ano)
    tmp.append(estado)
    values_ano_entidad = values_ano[values_ano["ENTIDAD"] == estado]
    for sexo in values_ano_entidad["SEXO"].unique():
      #print("Analizando a: " + sexo + "...")
      values_ano_entidad_sexo = values_ano_entidad[values_ano_entidad["SEXO"] == sexo]
      poblacion = values_ano_entidad_sexo["POBLACION"].sum()
      tmp.append(poblacion)
    mtrx.append(tmp)

array = np.array(mtrx)
result = pd.DataFrame(data = array, columns = columns)
result
  
```

```python
col_estados = df_H["EDO"].unique()
col_anos = []
for j in range(2000,2022):
  for i in range(len(col_estados)):
    col_anos.append(j)

homicidios = pd.DataFrame(index=col_anos, columns=["AÑO", "ENTIDAD", "HOMICIDIO_HOMBRES", "HOMICIDIO_MUJERES"])
homicidios["AÑO"] = col_anos

for i in col_estados:
  for j in range(2000, 2022):
    homicidios.loc[j, "ENTIDAD"] = i

homicidios
  

```
